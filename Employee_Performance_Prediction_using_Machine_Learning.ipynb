```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Uncleaned_employees_final_dataset (1).csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df["education"].fillna(df["education"].mode()[0], inplace = True)
```

```python
df["previous_year_rating"].fillna(df["previous_year_rating"].median(), inplace = True)
```

```python
## feature selection
input_features = ["department","region","education","recruitment_channel","no_of_trainings","age","previous_year_rating",
               "length_of_service","awards_won","avg_training_score"]
```

```python
X = df[input_features]
```

```python
output_feature = ["KPIs_met_more_than_80"]
```

```python
y = df[output_feature]
```

```python
## encoding
cat_col = ["department",	"region",	"education",	"recruitment_channel"]
```

```python
X = pd.get_dummies(X, columns=cat_col, drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
